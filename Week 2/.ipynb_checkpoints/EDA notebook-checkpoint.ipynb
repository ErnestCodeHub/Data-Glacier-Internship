# G2M insight for Cab Investment firm (EDA notebook)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

### Import Data

```python
cab_data = pd.read_csv('Cab_Data.csv')
city = pd.read_csv('City.csv')
transaction_id = pd.read_csv('Transaction_ID.csv')
customer_id = pd.read_csv('Customer_ID.csv')
```

```python
#Merge dataset 1 & 2
merge_df = cab_data.set_index('Transaction ID').join(transaction_id.set_index('Transaction ID'), how='outer')
```

```python
merge_df
```

```python
#Remove Indexing
merge_df.reset_index(inplace=True)
```

```python
#Add profit column 
merge_df['Profit'] = merge_df['Price Charged'] - merge_df['Cost of Trip']
```

```python
#Drop na
merge_df.dropna(inplace=True)

merge_df
#the rows has reduced from 440098 to 359392 
```

```python
#Merge dataset 2 & 3
merge_df1 = merge_df.set_index('Customer ID').join(customer_id.set_index('Customer ID'), how='left')
```

```python
#Remove Indexing
merge_df1.reset_index(inplace=True)
```

```python
merge_df1.dropna(inplace=True)

merge_df1
#the rows has reduced from 440098 to 359392
```

```python
# Remove commas from the 'Users' column & transform into string values
city['Users'] = city['Users'].astype(str).str.replace(',', '')
city['Population'] = city['Population'].astype(str).str.replace(',', '')

# Convert 'Users' and 'Population' columns to numeric type
city['Users'] = pd.to_numeric(city['Users'])
city['Population'] = pd.to_numeric(city['Population'])

# Calculate the percentage of users relative to the population
city['Percentage(%)'] = (city['Users'] / city['Population']) * 100

# Sort the DataFrame based on the newly calculated percentage column in ascending order
city.sort_values(by='Percentage(%)', ascending=False, inplace=True)

# Print the sorted DataFrame
print(city)
```

### Data Visualization

```python
sns.set(style="darkgrid")
```

```python
# Which city has the higher Taxi User/Population ratio?
plt.figure(figsize=(10, 6))

sns.barplot(data = city,x ="Percentage(%)", y = "City")

plt.title("Percentage (%) of Taxi Users compared to overall population")
plt.xlabel("Percentage (%)", fontsize=15)  # Increased x label size
plt.ylabel("City", fontsize=15)  # Increased y label size
plt.xticks(fontsize=10)  # Set x ticks size
plt.yticks(fontsize=10)  # Set y ticks size
plt.xlim(0, 45)  # Scale x to 50

plt.show()
```

```python
#which company has more transaction in each city
transactions_by_company_city = merge_df.groupby(['Company', 'City']).size().unstack()

# Plotting the transactions by company and city
transactions_by_company_city.plot(kind='bar', figsize=(12, 6))
plt.title('Number of Transactions by Company and City')
plt.xlabel('City')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=45)
plt.legend(title='Company')
plt.show()
```

```python
#which company earned more profit  
profit_by_company = merge_df.groupby('Company')['Profit'].sum()

# Plotting the total profit by company
profit_by_company.plot(kind='bar', figsize=(8, 6))
plt.title('Total Profit Earned by Company')
plt.xlabel('Company')
plt.ylabel('Total Profit')
plt.xticks(rotation=0)
plt.show()
```

```python
yellow_more_than_pink_cities = merge_df.loc[(merge_df['Company'] == 'Yellow Cab') & (merge_df['Company'] != 'Pink Cab'), 'City'].unique()
print("Cities where Yellow Cab has more transactions than Pink Cab:")
print(yellow_more_than_pink_cities)
```

```python
pink_more_than_yellow_cities = merge_df.loc[(merge_df['Company'] == 'Pink Cab') & (merge_df['Company'] != 'Yellow Cab'), 'City'].unique()
print("Cities where Pink Cab has more transactions than Yellow Cab:")
print(pink_more_than_yellow_cities)
```
